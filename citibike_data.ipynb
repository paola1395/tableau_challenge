```python
# Import dependencies
import pandas as pd
```

```python
# Import data files
april_2019 = "2019_citibike_data/april_2019.csv"
july_2019 = "2019_citibike_data/july_2019.csv"
oct_2019 = "2019_citibike_data/oct_2019.csv"

april_2020 = "2020_citibike_data/april_2020.csv"
july_2020 = "2020_citibike_data/july_2020.csv"
oct_2020 = "2020_citibike_data/oct_2020.csv"
```

```python
april_19 = pd.read_csv(april_2019, encoding="ISO-8859-1")
april_19.head()
```

```python
apr_years = pd.DataFrame(april_19["birth year"])
apr_years
```

```python
apr_age = pd.DataFrame(2019 - apr_years["birth year"])
apr_age
```

```python
apr_age_count = pd.DataFrame(apr_age['birth year']).value_counts()
apr_age_count.head(10)
```

```python
july_19 = pd.read_csv(july_2019, encoding="ISO-8859-1")
july_19.head()
```

```python
jul_years = pd.DataFrame(april_19["birth year"])
jul_age = pd.DataFrame(2019 - jul_years["birth year"])
jul_age_count = pd.DataFrame(jul_age['birth year']).value_counts()
jul_age_count.head(10)
```

```python
apr_jul = pd.concat([april_19, july_19])
apr_jul.head()
```

```python
oct_19 = pd.read_csv(oct_2019, encoding="ISO-8859-1")
oct_19.head()
```

```python
oct_years = pd.DataFrame(oct_19["birth year"])
oct_age = pd.DataFrame(2019 - oct_years["birth year"])
oct_age_count = pd.DataFrame(oct_age['birth year']).value_counts()
oct_age_count.head(10)
```

```python
combined_19 = pd.concat([apr_jul, oct_19])
combined_19.head()
```

```python
april_20 = pd.read_csv(april_2020, encoding="ISO-8859-1")
april_20.head()
```

```python
july_20 = pd.read_csv(july_2020, encoding="ISO-8859-1")
july_20.head()
```

```python
aprjul_20 = pd.concat([april_20, july_20])
aprjul_20.head()
```

```python
oct_20 = pd.read_csv(oct_2020, encoding="ISO-8859-1")
oct_20.head()
```

```python
combined_20 = pd.concat([aprjul_20, oct_20])
combined_20.head()
```

```python
# Combine all of the combined DFs
citibike_data = pd.concat([combined_19, combined_20])
citibike_data.head()
```

```python
# Export to csv
citibike_data.to_csv("citibike_data.csv", index=False)
```
